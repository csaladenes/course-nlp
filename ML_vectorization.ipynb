```python
import numpy as np
from time import monotonic
import matplotlib.pyplot as plt
```

```python
W=np.random.normal(0,1,10000)
print(W.shape)
W[10:100]
```

```python
plt.hist(W, 100, density = 1, color ='green',alpha = 0.7)
plt.show()
```

```python

def non_vectorized_sum(numbers):
  sum=0
  for num in numbers:
    sum+=num
  return sum
%timeit non_vectorized_sum(W)
```

```python
start = monotonic()
sum1 = non_vectorized_sum(W)
end = monotonic()
timediff1 = end-start
print(f'Time diff: {timediff1}, sum: {sum1}')
```

Trick to make this sum operation a matrix multiplication?

```python

def vectorized_sum(numbers):
  ones = np.ones(numbers.shape)
  sum = ones.T.dot(numbers)
  return sum
%timeit vectorized_sum(W)
```

```python
start = monotonic()
sum2 = vectorized_sum(W)
end = monotonic()
timediff2 = end-start
print(f'Time diff: {timediff2}, sum: {sum2}')
```
